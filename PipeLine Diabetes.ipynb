```python
X=df.iloc[:,:-1]
X.head()
```

```python
y=df['Outcome']
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
numerical_col=[i for i in X.columns if X[i].dtype!='categorical']
numerical_col
```

```python
categorical_col=[i for i in X.columns if X[i].dtype=='categorical']
categorical_col
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipeline=Pipeline(
    steps=[
        ('scaler',StandardScaler())
    ]
)

cat_pipeline=Pipeline(
    steps=[
    ('onehotencoding',OneHotEncoder())
    ]
)
```

```python
preprocessing=ColumnTransformer([
    ('num_pipeline',num_pipeline,numerical_col),
    ('cat_pipeline',cat_pipeline,categorical_col)
])
```

```python
X_train=preprocessing.fit_transform(X_train)
X_test=preprocessing.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report
```

```python
models={
    'Logistic Regression': LogisticRegression(),
    'SVC': SVC(),
    'Decision tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier()
}
```

```python
def evaluate_model(X_train,X_test,y_train,y_test):
    report={}
    classification_rpt={}
    for i in range(len(models)):
        model=list(models.values())[i]
        model.fit(X_train,y_train)
        y_pred=model.predict(X_test)
        test_score=accuracy_score(y_test,y_pred)
        report[list(models.keys())[i]]=test_score
        classification_rpt[list(models.keys())[i]]=classification_report(y_test,y_pred)
    return report,classification_rpt
```

```python
report,classification_rpt=evaluate_model(X_train,X_test,y_train,y_test)
```
